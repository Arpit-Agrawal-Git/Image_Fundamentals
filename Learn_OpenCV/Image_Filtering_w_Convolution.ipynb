```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sample_pic.png')
# image - cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(image[:,:,::-1])
```

### Identity kernel

Centre 1 , others 0

```python
kernel = np.array([[0,0,0],
                   [0,1,0],
                   [0,0,0]])

identity = cv2.filter2D(image, ddepth=-1, kernel=kernel)

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(identity[:,:,::-1])
plt.title("Identity")
plt.axis("off")
```

Kernel : BLURR

```python
kernel = np.ones((3,3), np.float32)/9

img = cv2.filter2D(image, ddepth=-1, kernel=kernel)

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(img[:,:,::-1])
plt.title("Blurred Image")
plt.axis("off")
```

cv2 blur method

```python
img_blur = cv2.blur(image, ksize=(5,5))

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(img_blur[:,:,::-1])
plt.title("Blurred Image")
plt.axis("off")
```

cv2 Gaussian Blur

```python
img_gb = cv2.GaussianBlur(image, ksize=(5,5), sigmaX=0)

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(img_gb[:,:,::-1])
plt.title("Gaussian Blurred Image")
plt.axis("off")
```

Median Blurr

```python
img_mb = cv2.medianBlur(image, ksize=25)

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(img_blur[:,:,::-1])
plt.title("Median Blurred Image")
plt.axis("off")
```

Sharpen Blurr

```python
kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharpen = cv2.filter2D(image, ddepth=-1, kernel=kernel)

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(sharpen[:,:,::-1])
plt.title("Sharperned Image")
plt.axis("off")
```

Bilateral Filter

Removes noise

```python
bf = cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)

plt.figure(figsize=(12,10))
plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Original")
plt.axis("off")

plt.figure(figsize=(12,10))
plt.subplot(122)
plt.imshow(bf[:,:,::-1])
plt.title("Bilateral Filter")
plt.axis("off")
```
