```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sample_pic.png')
```

```python
height, width = image.shape[:2]
```

Translation : Moving image linearly

First require Translation Matrix

For instance : moving quarter way down

```python
tx, ty = width/4, height/4
```

```python
translation_matrix = np.array([
  [1,0,tx],
  [0,1,ty]
], dtype=np.float32)
```

### Components of the Matrix

- 1 in the first row, first column and second row, second column represents the scaling factor for the x and y coordinates, respectively. Since they are 1, it means there's no scaling involved.
- 0 in the first row, second column and second row, first column represents the rotation factor. Since they are 0, it means there's no rotation involved.
- tx in the first row, third column represents the translation distance along the x-axis.
- ty in the second row, third column represents the translation distance along the y-axis.

```python
translated_image = cv2.warpAffine(src=image, M=translation_matrix,dsize=(width,height))
```

```python
plt.axis('off')
plt.imshow(translated_image[:,:,::-1])
```
