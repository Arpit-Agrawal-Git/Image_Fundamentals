```python
import cv2
import numpy as np
```

```python
# Create video capture object
video = cv2.VideoCapture('sample_video.mp4')
```

```python
if(video.isOpened()):
  print("Yes Yes")
else:
  print("No No")
```

```python
fps = video.get(cv2.CAP_PROP_FPS)
frame_count = video.get(cv2.CAP_PROP_FRAME_COUNT)
width = video.get(cv2.CAP_PROP_FRAME_WIDTH)
height = video.get(cv2.CAP_PROP_FRAME_HEIGHT)
```

```python
print(fps)
print(frame_count)
print(width)
print(height)
```

```python
while(video.isOpened()):
  ret,frame = video.read()
  if ret == True:
    cv2.imshow('Video Frame', frame)
    key = cv2.waitKey(10)
    if key == ord('q'):
      break
  else:
    break
```

```python
video.release()
cv2.destroyAllWindows()
```

### Writing webcam capture to video file

```python
video_write = cv2.VideoCapture(1)
```

```python
width = int(video_write.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video_write.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
frame_size = (width, height)
```

### Need to define name of video, fourcc, fps and frame size

```python
video_out = cv2.VideoWriter('webcam_gray.mp4', cv2.VideoWriter_fourcc(*'mp4v'),32,frame_size)
```

```python
while(video_write.isOpened()):
  ret,frame = video_write.read()
  if ret == True:
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    frame_stack = np.stack([frame]*3, axis = -1)
    video_out.write(frame_stack)
    cv2.imshow('Video Frame', frame)
    key = cv2.waitKey(10)
    if key == ord('q'):
      break
  else:
    break
```

```python
video_write.release()
video_out.release()
cv2.destroyAllWindows()
```
