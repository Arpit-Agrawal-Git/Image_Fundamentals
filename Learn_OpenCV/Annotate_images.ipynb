```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sample_pic.png')
```

For Ellipse arguments : image, ellipse center co-ord, axis length, angle of ellipse, start angle, end angle, color, thickness

-1 is to fill inside

```python
cv2.ellipse(image, (550,200), (50,50),0,180,360, (0,255,255),-1)
```

```python
plt.axis('off')
plt.imshow(image[:,:,::-1])
```

For Line arguments : image, start co-ord, end co-ord, color and thickness

```python
cv2.line(image, (600,470), (750,200), (36,249,47), 15)
```

```python
plt.axis('off')
plt.imshow(image[:,:,::-1])
```

For circle arguments : image, center co-ord, radius in px, color, thickness

-1 to fill the circle

```python
cv2.circle(image, (300,200), 50, (0,255,255), -1)
```

```python
plt.axis('off')
plt.imshow(image[:,:,::-1])
```

For Rectangle arguments : image, start co-ord, end co-ord, color and thickness

```python
cv2.rectangle(image,(700,700),(1410,800),(0,255,255),5)
```

```python
plt.axis('off')
plt.imshow(image[:,:,::-1])
```

For Text : image, text, start co-ord of text, font, font size, font color, thickness

```python
cv2.putText(image,"That 70s show", (710,790),cv2.FONT_HERSHEY_SIMPLEX,3,(0,255,255),5)
```

```python
plt.axis('off')
plt.imshow(image[:,:,::-1])
```
