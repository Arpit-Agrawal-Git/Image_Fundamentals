```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Image Thresholding :

- separate object from background
- classify pixel as either foreground or background


```python
image = cv2.imread('sample_pic.png',0)
th, res = cv2.threshold(image, 130, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10,8))
plt.subplot(121)
plt.axis('off')
plt.title("Original Image")
plt.imshow(image[:,:])

plt.figure(figsize=(10,8))
plt.subplot(122)
plt.axis('off')
plt.title("Threshold Image")
plt.imshow(res[:,:])
```

```python
image = cv2.imread('sample_pic.png', 0)
th, res = cv2.threshold(image, 130, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(10,8))
plt.subplot(121)
plt.axis('off')
plt.title("Original Image")
plt.imshow(image[:,:])

plt.figure(figsize=(10,8))
plt.subplot(122)
plt.axis('off')
plt.title("Threshold Image Inverse")
plt.imshow(res[:,:])
```
