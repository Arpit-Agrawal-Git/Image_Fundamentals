```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

input = source image

output = destination image

Read with cv2 and display with plt(for display in notebooks)

```python
image = cv2.imread('sample_pic.png')
```

```python
plt.imshow(image[:,:,::-1])
```

```python
height, width = image.shape[:2]
aspect_ratio = width/height
print(f"Width : {width}")
print(f"Height : {height}")
print(f"Aspect Ratio : {aspect_ratio}")
```

## Downsizing with destination size

```python
down_width = 300
down_height = 200
down_points = (down_width,down_height)
resized_down = cv2.resize(image, down_points, interpolation=cv2.INTER_LINEAR)
```

## Upsizing with destination size

```python
up_width = 1920
up_height = 1080
up_points = (up_width,up_height)
resized_up = cv2.resize(image, up_points, interpolation=cv2.INTER_LINEAR)
```

```python
plt.figure(figsize=[15,15])
plt.subplot(131); plt.imshow(image[:,:,::-1]); plt.title("Original Image\nAspect Ratio : 1.5")
plt.subplot(132); plt.imshow(resized_down[:,:,::-1]); plt.title("Resized Down Image\nAspect Ratio : 1.5")
plt.subplot(133); plt.imshow(resized_up[:,:,::-1]); plt.title("Resized up Image\nAspect Ratio : 1.5")
```

## Using Scaling Factors

```python
# Scaling Up factor
scale_up_x = 1.2
scale_up_y = 1.5

# Scaling down factor
scale_down = 0.5

scale_f_down = cv2.resize(image, None, fx=scale_down, fy=scale_down, interpolation=cv2.INTER_LINEAR)
scale_f_up = cv2.resize(image, None, fx=scale_up_x, fy=scale_up_y, interpolation=cv2.INTER_LINEAR)
```

```python
plt.figure(figsize=[15,15])
plt.subplot(121); plt.imshow(scale_f_down[:,:,::-1])
plt.title("Scaled down image, size = {}".format(scale_f_down.shape[:2]))
plt.subplot(122); plt.imshow(scale_f_up[:,:,::-1])
plt.title("Scaled up image, size = {}".format(scale_f_up.shape[:2]))
```
