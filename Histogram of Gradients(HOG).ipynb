```python
from skimage import feature, io, color, exposure

# Load an image (you'll need to provide the path to your image file)
image = io.imread(r"C:\Users\arpit\Pictures\Screenshot 2023-03-22 120656.jpg")
image_gray = color.rgb2gray(image)

# Set the parameters for HOG
pixels_per_cell = (16, 16)  # size of a cell (in pixels)
cells_per_block = (1, 1)    # number of cells per block
orientations = 8          # number of bins

# Compute HOG features
hog_features, hog_image = feature.hog(image_gray, orientations=orientations,
                                      pixels_per_cell=pixels_per_cell,
                                      cells_per_block=cells_per_block,
                                      visualize=True, feature_vector=True)

hog_image_r = exposure.rescale_intensity(hog_image,in_range=(0,10))
# hog_features contains the feature vector of the HOG descriptor
# hog_image is an image suitable for visualizing the histogram of oriented gradients

# If you want to display the HOG image
import matplotlib.pyplot as plt
plt.imshow(hog_image_r, cmap='gray')
plt.show()
```
