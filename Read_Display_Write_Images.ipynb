```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

imread() modes
- IMREAD_COLOR or 1
- IMREAD_UNCHANGED or -1
- IMREAD_GRAYSCALE or 0

OpenCV reads image in BGR format

```python
image = cv2.imread('sample_pic.png',1)
```

```python
# cv2.imshow('window',image)
# cv2.waitKey(0)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off') # to hide axis labels
plt.show()
```
