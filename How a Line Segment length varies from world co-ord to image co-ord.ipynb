```python
from math import sqrt
import matplotlib.pyplot as plt
```

```python
X1 = -5
X2 = 5
f = 1 #focal length

L = []

for Z in range(10,1000):
    x1 = -f*X1/Z
    x2 = -f*X2/Z
    L.append(sqrt((x1-x2)**2))

plt.plot(L)
plt.xlabel('distance(Z)')
plt.ylabel('Length')
```

This demos how a 1D line in world coordinate system changes in image cordinates. This is fundamental concept behind perspective projection that explains why objects farther off from camera appear smaller
