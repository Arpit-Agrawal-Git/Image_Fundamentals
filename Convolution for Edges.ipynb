```python
import numpy as np
from scipy.ndimage import convolve
from scipy import ndimage
import matplotlib.pyplot as plt
import imageio

# Load an example image
image = imageio.imread(r"C:\Users\arpit\Pictures\Screenshots\Screenshot 2023-04-09 211822.png", pilmode='L')  # Load as grayscale

# Define the kernels for horizontal and vertical derivatives
kernel_fx = np.array([[1, 0, -1],
                      [2, 0, -2],
                      [1, 0, -1]])

kernel_fy = np.array([[1,  2,  1],
                      [0,  0,  0],
                      [-1, -2, -1]])

# Convolve the image with the kernels
fx = convolve(image, kernel_fx)
fy = convolve(image, kernel_fy)

# Compute the gradient magnitude
gradient_magnitude = np.sqrt(fx*2 + fy*2)

# Display the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title('Horizontal Derivative fx')
plt.imshow(fx, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Vertical Derivative fy')
plt.imshow(fy, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Gradient Magnitude')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.show()
```
