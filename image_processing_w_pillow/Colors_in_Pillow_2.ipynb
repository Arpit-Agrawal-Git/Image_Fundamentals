```python
from PIL import Image, ImageFilter
from numpy import array
```

```python
image = Image.open('sample_pic.png')
```

## Palette

```python
palette = image.convert('P')
```

```python
palette
```

256 colors in this image

```python
print(array(palette))
```

The values here are not color but reference to a color in a palette

```python
print(palette.getpalette())
```

```python
print(array(palette.getpalette()).reshape(256,3))
```

These are RGB tuples

## Limiting the number of colors in a palette

```python
palette_16 = image.convert('P', palette = Image.Palette.ADAPTIVE, colors = 16)
```

```python
palette_16
```

### Update color palette

```python
new_palette = [x//2 for x in palette_16.getpalette()]
palette_16.putpalette(new_palette)
```

```python
palette_16
```

Brightness of all pixel is halved
