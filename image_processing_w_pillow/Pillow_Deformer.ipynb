```python
from PIL import Image, ImageOps
```

```python
class Deformer():
  def getmesh(self,img):
    w,h = img.size
    # define a shape in the OG image
    source_shape = (0,0,0,h,w-1000,h,w,0)
    
    # define a rectange in the new image
    target_rect = (0,0,w//2,h) # left, top, right, bottom
    
    # return all shapes
    return [(target_rect, source_shape)]
```

```python
# Image import
image = Image.open('sample_pic.png')
```

```python
Deform = ImageOps.deform(image, Deformer())
```

```python
Deform
```

```python
class Deformer1():
  def getmesh(self,img):
    w,h = img.size
   
    left = ((0,0,w//2,h),(0,0,0,h,w//2,h,w//2,0))
    right = ((w//2,0,w,h),(w//2,0,w//2,h,0,h,0,0)) # left, top, right, bottom
    
    # return all shapes
    return [left,right]
```

```python
# Image import
image = Image.open('sample_pic.png')
```

```python
Deform1 = ImageOps.deform(image, Deformer1())
```
